```python
# Original problem

import random
import matplotlib.pyplot as plt

def simulate_trial(max_steps=1000):
    position = 1
    landed_steps = set()
    landed_steps.add(position)
    for _ in range(max_steps):
        step = random.choice([1, 2])
        position += step
        landed_steps.add(position)
        if position > 100:
            break
    return landed_steps

def run_trials(num_trials=100):
    step_counts = {}
    for _ in range(num_trials):
        landed_steps = simulate_trial()
        for step in landed_steps:
            if step in step_counts:
                step_counts[step] += 1
            else:
                step_counts[step] = 1
    return step_counts

def plot_results(step_counts):
    steps = sorted(step_counts.keys())
    counts = [step_counts[step] for step in steps]
    
    plt.bar(steps, counts, color='blue')
    plt.xlabel('Step Number (n)')
    plt.ylabel('Number of Trials Landing on Step n')
    plt.title('Number of Trials Landing on Each Step (10000 Trials)')
    plt.show()
    
step_counts = run_trials(num_trials=10000)

plot_results(step_counts)
```

```python
# Three step recursion

import random
import matplotlib.pyplot as plt

def simulate_trial(max_steps=1000):
    position = 1
    landed_steps = set()
    landed_steps.add(position)
    for _ in range(max_steps):
        step = random.choice([1, 2, 3])
        position += step
        landed_steps.add(position)
        if position > 100:
            break
    return landed_steps

def run_trials(num_trials=100):
    step_counts = {}
    for _ in range(num_trials):
        landed_steps = simulate_trial()
        for step in landed_steps:
            if step in step_counts:
                step_counts[step] += 1
            else:
                step_counts[step] = 1
    return step_counts

def plot_results(step_counts):
    steps = sorted(step_counts.keys())
    counts = [step_counts[step] for step in steps]
    plt.bar(steps, counts, color='blue')
    plt.xlabel('Step Number (n)')
    plt.ylabel('Number of Trials Landing on Step n')
    plt.title('Number of Trials Landing on Each Step (10000 Trials, Three-Sided Coin)')
    plt.show()

step_counts = run_trials(num_trials=10000)

plot_results(step_counts)
```
